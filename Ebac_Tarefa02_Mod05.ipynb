01/02- Marcar Default

```python
pagamentos = pd.read_csv('pagamentos_largo.csv')
pagamentos
```

Buscando por clientes 'Default'

```python
pagamentos_tratados = pagamentos.fillna(0)
pagamentos_tratados['credito'] = 0

for i in pagamentos_tratados.columns[:-1]:
  pagamentos_tratados.loc[(pagamentos_tratados[i].isin(['0', '1', 'C'])), 'credito'] = 'nao_default'

for i in pagamentos_tratados.columns[:-1]:
  pagamentos_tratados.loc[(pagamentos_tratados[i].isin(['2', '3', '4', '5'])), 'credito'] = 'default'

pagamentos_tratados[pagamentos_tratados['credito'] == 'default']
```

03/04- Marcando proponentes com risco juntando informações

```python
propostas = pd.read_csv('application_record.csv')
propostas
```

```python
propostas_avaliacao = pd.merge(propostas, pagamentos_tratados[['ID', 'credito']], how='left', on='ID')
propostas_default = propostas_avaliacao[propostas_avaliacao['credito'] == 'default']
```
